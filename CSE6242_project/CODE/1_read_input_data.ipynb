1.2 remove 1st observation by nh

```python
df_nh_data = df_nh_data.sort_values(by=['nh_id','date'],ignore_index=True)
df_nh_data['num_obs_nh'] = df_nh_data.groupby('nh_id').cumcount()
df_nh_data = df_nh_data[df_nh_data['num_obs_nh'] > 0]
df_nh_data.drop(columns = ['num_obs_nh'],inplace = True)
df_nh_data.reset_index(drop=True,inplace = True)
```

1.3 combine treatment variables

```python
cols = [col for col in list(df_nh_data.columns) if "n_treatment" in col]
#df_nh_data[cols]=df_nh_data[cols].fillna(0)
df_nh_data["num_treatment"] = df_nh_data[cols].sum(axis=1, min_count=1)
df_nh_data.drop(columns = cols, inplace = True)
```

1.4 combine shortage of staff variabels

    if all 4 vars nan -> nan

    if 1 or more valid value, all N -> N

    if 1 or more valid value, 1 or more is Y -> Y

```python
cols = ["shortage_nursing_staff","shortage_clincal_staff","shortage_aides","shortage_other_staff"]
df_nh_data["shortage_staff"] = df_nh_data[cols].eq("Y").any(axis=1)
df_nh_data.loc[df_nh_data[cols].isnull().all(axis = 1),"shortage_staff"] = np.NaN
#df_nh_data.loc[(df_nh_data['shortage_aides'] == "Y") & (df_nh_data['shortage_clincal_staff'].isnull())].head()
df_nh_data.drop(columns = cols, inplace = True)
```

2.1.1 keep relevant fields only

2.2.0 read ltc data

2.2.1 keep relevant columns only and rename them

PROV1680: facility id

PROV0475: facility name

acuindex2: Acuity Index is a measure of the care needed by a nursing home's residents. 

agghighcfs: Proportion of residents admitted during the calendar year with a Cognitive Function Scale (CFS) score of 4 (severe cognitive impairment).





```python
df_nh_char_ltc_column_dict = {'PROV1680':'nh_id',
                              #'PROV0475':'nh_name',
                              'acuindex2':'acuindex2',
                              'paymcaid':'paymcaid',
                              "paymcare":"paymcare",
                              "agghighcfs":"agghighcfs",
                              "pctfem":"pctfem",
                              "pctunder65":"pctunder65",
                              "pctblack_mds3":"pctblack_mds3",
                              "pcthisp_mds3":"pcthisp_mds3",
                              'pctwhite_mds3':'pctwhite_mds3',
                              }
df_nh_char_ltc.rename(columns=df_nh_char_ltc_column_dict,inplace=True)
df_nh_char_ltc = df_nh_char_ltc[list(df_nh_char_ltc_column_dict.values())]
```

2.3.0 merge df_nh_char and df_nh_char_ltc

```python
df_nh_char = pd.merge(df_nh_char,df_nh_char_ltc,how="left",on=["nh_id"])
```

2.3.1 convert numeric cols to numeric data type, replace invalid values with np.NaN

```python
cols_to_num = ['acuindex2','paymcaid','paymcare','pctfem','agghighcfs','pctunder65','pctblack_mds3','pcthisp_mds3','pctwhite_mds3']
df_nh_char[cols_to_num] = df_nh_char[cols_to_num].apply(pd.to_numeric,errors='coerce')
```

```python
lat_long=pd.read_csv(filepath+'data/nursing_home_lat_long.csv',dtype = {'Federal Provider Number':'str'})
df_nh_char = pd.merge(df_nh_char,lat_long,how="left",left_on=["nh_id"],right_on=['Federal Provider Number'])
df_nh_char.drop(columns = ['Federal Provider Number'],inplace = True)
```
